```python
#Import winter data
winter_data = pd.read_csv("MAP_Winter_2020-21.csv")
```

```python
#Import student demographics data
stu_demo_data = pd.read_csv("Student_Demographics_2020-21.csv")
```

```python
#Look at first 5 rows
stu_demo_data.head()
```

<font size="4"><b>Data Cleansing</b></font>

```python
#Concatenate fall data and winter data to create a single data frame
combine = fall_data.append(winter_data)
```

```python
combine
```

```python
#Join student demographic data to MAP data and limit to most important features
join_stu_demo = combine[["Fake_Student_ID","TermName","SchoolName",
                        "Grade","TestRITScore","TestPercentile","PercentCorrect"]].merge(stu_demo_data[["Fake_Student_ID","Race","Ethnicity","Sex","LEP",
                                                                                                       "HomeLanguage","SpEdFlag","TAGFlag",
                                                                                                       "Econ_Disadvantage","Migrant","DyslexiaFlag"]], 
                                                                                         on = "Fake_Student_ID", how = "inner")
```

```python
join_stu_demo
```

```python
new_data = join_stu_demo
```

```python
#Check for missing values
new_data.isnull().sum()
```

```python
#HomeLanguage is the only feature that has missing values. For the purpose of this project, I am going to drop these rows.
#Under normal circumstances, using my access to student info would allow me to fill these empty values with the correct info
#for that student.
new_data.dropna(inplace = True)
```

```python
#Verify rows with missing values have been dropped
new_data.isnull().sum()
```
