```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import fall data
fall_data = pd.read_csv("MAP_Fall_2020-21.csv")
```

```python
#Look at first 5 rows
fall_data.head()
```

```python
#Import winter data
winter_data = pd.read_csv("MAP_Winter_2020-21.csv")
```

```python
#Look at first 5 rows
winter_data.head()
```

```python
#Import student demographics data
stu_demo_data = pd.read_csv("Student_Demographics_2020-21.csv")
```

```python
#Look at first 5 rows
stu_demo_data.head()
```

```python
#Concatenate fall data and winter data to create a single data frame
combine = fall_data.append(winter_data)
```

```python
combine
```

```python
#Join student demographic data to MAP data and limit to most important features
join_stu_demo = combine[["Fake_Student_ID","TermName","SchoolName",
                        "Grade","Subject","TestRITScore","TestPercentile","PercentCorrect"]].merge(stu_demo_data[["Fake_Student_ID","Race","Ethnicity","Sex","LEP",
                                                                                                       "HomeLanguage","SpEdFlag","TAGFlag",
                                                                                                       "Econ_Disadvantage","Migrant","DyslexiaFlag"]], 
                                                                                         on = "Fake_Student_ID", how = "inner")
```

```python
join_stu_demo
```

```python
new_data = join_stu_demo
```

```python
#Check for missing values
new_data.isnull().sum()
```

```python
#HomeLanguage is the only feature that has missing values. For the purpose of this project, I am going to drop these rows.
#Under normal circumstances, using my access to student info would allow me to fill these empty values with the correct info
#for that student.
new_data.dropna(inplace = True)
```

```python
#Verify rows with missing values have been dropped
new_data.isnull().sum()
```

```python
new_data
```

```python
#Change values of fields where applicable to make data easily understood by users/presenters
new_data['Ethnicity'] = new_data['Ethnicity'].replace(['H'],'Hispanic')
new_data['Ethnicity'] = new_data['Ethnicity'].replace(['N'],'Not Hispanic')

new_data['LEP'] = new_data['LEP'].replace(['T'],'LEP')
new_data['LEP'] = new_data['LEP'].replace(['F'],'Not LEP')

new_data['SpEdFlag'] = new_data['SpEdFlag'].replace(['T'],'SpEd')
new_data['SpEdFlag'] = new_data['SpEdFlag'].replace(['F'],'Not SpEd')

new_data['TAGFlag'] = new_data['TAGFlag'].replace(['T'],'T&G')
new_data['TAGFlag'] = new_data['TAGFlag'].replace(['F'],'Not T&G')

new_data['Econ_Disadvantage'] = new_data['Econ_Disadvantage'].replace(['T'],'Eco Dis')
new_data['Econ_Disadvantage'] = new_data['Econ_Disadvantage'].replace(['F'],'Not Eco Dis')

new_data['Migrant'] = new_data['Migrant'].replace(['Y'],'Migrant')
new_data['Migrant'] = new_data['Migrant'].replace(['N'],'Not Migrant')

new_data['DyslexiaFlag'] = new_data['DyslexiaFlag'].replace(['T'],'Dyslexic')
new_data['DyslexiaFlag'] = new_data['DyslexiaFlag'].replace(['F'],'Not Dyslexic')
```

```python
new_data.head()
```

```python
#Clean up field names to make data easily understood by users/presenters
new_data.rename(columns={'Fake_Student_ID':'Student_ID',
                          'TermName':'Term',
                          'SchoolName':'School',
                          'TestRITScore':'RIT Score',
                          'TestPercentile':'Percentile',
                          'LEP':'LEP Flag',
                          'TAGFlag':'T&G Flag',
                          'Econ_Disadvantage':'Eco Dis Flag',
                          'Migrant':'Migrant Flag',
                          'DyslexiaFlag':'Dyslexia Flag'}, 
                 inplace=True)
```

```python
new_data.head()
```

```python
new_data.rename(columns={'Student_ID':'Student ID'},inplace = True)
```

```python
new_data.head()
```

```python
new_data['Race'] = new_data['Race'].replace(['M Multi-Racial'],'Multi-Racial')
new_data['Race'] = new_data['Race'].replace(['P Native Hawaiian or Pacific'],'Native Hawaiian or Pacific')
```
